```python
import matplotlib.pyplot as plt
def show_slices(slices):
   """ Function to display row of image slices """
   fig, axes = plt.subplots(1, len(slices))
   for i, slice in enumerate(slices):
       axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
slice_0 = epi_img_data[144, :, :]
slice_1 = epi_img_data[:, 160, :]
slice_2 = epi_img_data[:, :, 99]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for EPI image")
```

```python
n_i, n_j, n_k = epi_img_data.shape
center_i = (n_i - 1) // 2  # // for integer division
center_j = (n_j - 1) // 2
center_k = (n_k - 1) // 2
print(center_i, center_j, center_k)
center_vox_value = epi_img_data[center_i, center_j, center_k]
print(center_vox_value)
```
